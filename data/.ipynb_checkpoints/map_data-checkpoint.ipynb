```python
import os
import matplotlib.pyplot as plt
import math

%matplotlib inline

filename = 'map_data.txt'
fo = open(filename, "r")

x = []
y = []
landmark_ID = []

for line in fo:
    splitted_line = line.split('\t')
    x.append(float(splitted_line[0]))
    y.append(float(splitted_line[1]))
    landmark_ID.append(int(splitted_line[2].split('\n')[0])) #on code starting at 0
fo.close()

#print(x)
#print(y)
#print(landmark_ID)


# plot the chart
plt.figure(figsize=(20,10))

#closest_landmark = [2, 38, 31, 39, 39, 37, 39, 16, 39, 16, 31]
closest_landmark = landmark_ID

plt.scatter(x,y)
for i in range(len(x)):

    # this method is called for each point
    plt.annotate(landmark_ID[i], # this is the text
                 (x[i],y[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

particle = [1.038598e+02,-2.591200e+01,3.429132e+00]
particle_x = particle[0]
particle_y = particle[1]
particle_theta = particle[2]
plt.scatter(particle_x,particle_y,c='r')


obs_map_x = [9.073868e+01,1.018101e+02,1.357554e+02,5.930588e+01]
obs_map_y = [-3.357570e+01,5.298689e+00,-1.453659e+01,-4.529769e+01]
obs_map_label = [0,1602245632,1602245632,1602245632]
plt.scatter(obs_map_x,obs_map_y,c='y')

for i in range(len(obs_map_x)):
    plt.annotate(obs_map_label[i], # this is the text
                 (obs_map_x[i],obs_map_y[i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(-10,0), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
                 



plt.show()
```
